```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
data1=pd.read_csv("data (1).csv")
```

```python
data1.head()
```

```python
data1.info()
```

```python
data1.isnull().sum()
```

```python
data1=data1.drop(["Unnamed: 32","id"],axis=1)
```

```python
data1["diagnosis"] = data1["diagnosis"].replace({"M": 1, "B": 0})
```

```python
data1.diagnosis
```

```python
y=data1["diagnosis"]
```

```python
X=data1.drop(["diagnosis"],axis=1)
```

```python
X_train, X_test, y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
svc_model= SVC(kernel="linear")
```

```python
svc_model.fit(X_train,y_train)
```

```python
y_pred=svc_model.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
svc_params={"C":np.arange(1,10)}
```

```python
svc_cv_model=GridSearchCV(svc_model,svc_params,cv=10).fit(X_train,y_train)
```

```python
a=svc_cv_model.best_params_
```

```python
svc_tuned=SVC(kernel="linear",C=3).fit(X_train,y_train)
```

```python
svc_tuned.coef_
```

```python
y_pred=svc_tuned.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```
